2. Lépés
Minden sor egy terméket jelent. Most soronként (cikkenként) a következő 12 hónap előrjelzését összeadom, és kiegészítem az adatbázisamat ezen értékekkel az utolsó sorban. Az oszlopot Összegnek nevezem el. 

```python

FCCurrentlist = ['Május','Június','Július','Augusztus','Szeptember','Október','November','December','Január','Február','Március','Április']
FCCurrent['Összeg'] = FCCurrent[FCCurrentlist].sum(axis=1)
FCCurrent.head(30)
```

```python
FCPrevlist = ['Május','Június','Július','Augusztus','Szeptember','Október','November','December','Január','Február','Március','Április']
FCPrev['Összeg'] = FCPrev[FCPrevlist].sum(axis=1)
FCPrev.head(27)
```

3.Lépés Létrehozok egy külön adatbázist ami tartalmazni fogja az összes cikkszámot, illetve külön oszlopba rendezve az aktuális és az előző előrejelzések adatait.

```python
current_df = pd.DataFrame(FCCurrent)  
prev_df = pd.DataFrame(FCPrev)  
prev_df_subset = prev_df[['EKS']]

#Mivel a sorok nem megegyezőek a két táblázatban, össze kell fésülni azokat. Ezt merge-éssel lehet megoldani
merged_df = pd.merge(current_df, prev_df_subset, left_on='EKS', right_on='EKS')

# Az új táblázat (Összegzés) 

Összegzés_df = pd.DataFrame({
    'EKS': current_df.iloc[:, 0],
    'Name': current_df.iloc[:, 1],
    'CurrentFC': current_df.iloc[:, 14], # Mindkét táblázatban a 16. oszlop jelenti az összegzést
    'PrevFC': prev_df.iloc[:, 16]  # Mindkét táblázatban a 16. oszlop jelenti az összegzést
})

# Az új táblázat megjelenítése
Összegzés_df.head(27)
```

4. Lépés
Adatok összefésülése után, a 2. lépésekben látható módszerrel. Kiegészítem az Összegzés táblát egy olyan oszloppal, ami megmutatja a különbségeket.

```python

Összegzés_df['Diff'] = Összegzés_df['CurrentFC'] - Összegzés_df['PrevFC']
Összegzés_df.head(27)
```

```python
Összegzés_df = Összegzés_df.sort_values(by='Diff')
Összegzés_df.head(274)
```
